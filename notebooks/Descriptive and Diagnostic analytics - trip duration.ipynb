### How much time has it been taken to go from some place to another place

```python
df = dataset_raw[["pickup_latitude","pickup_longitude","dropoff_latitude","dropoff_longitude","Pick_ID","Drop_ID","month","day","hour","minute","weekday","trip_duration"]] 
train,test = train_test_split(df, test_size=0.2, random_state=42)
```

#### 1. Duration estimate by location using six months data

```python
estimated_df = train.groupby(["Pick_ID","Drop_ID"]).agg({"trip_duration":["mean"]}).reset_index()
```

```python
result = test.merge(estimated_df,on=['Pick_ID','Drop_ID'],how='left')
result.rename(columns = {('trip_duration', 'mean'):'trip_duration_mean'}, inplace = True) 
result["error"] = abs(result["trip_duration"] - result["trip_duration_mean"])
```

```python
rmse = ((result["error"]**2).mean() ** 0.5)
print("RMSE in minutes :",rmse/60)
```

```python
result["trip_duration_mean"].isna().sum() *100 / len(result)
```

```python
result[0:10]
```

#### 2. Duration estimate by location using six months data with time factor

```python
estimated_df = train.groupby(["Pick_ID","Drop_ID","weekday","hour"]).agg({"trip_duration":["mean"]}).reset_index()
```

```python
result = test.merge(estimated_df,on=['Pick_ID','Drop_ID','weekday','hour'],how='left')
result.rename(columns = {('trip_duration', 'mean'):'trip_duration_mean'}, inplace = True) 
result["error"] = abs(result["trip_duration"] - result["trip_duration_mean"])
```

```python
rmse = ((result["error"]**2).mean() ** 0.5)
print("RMSE in minutes :",rmse/60)
```

```python
result["trip_duration_mean"].isna().sum() *100 / len(result) 
```

```python
def route_dis(loc1,loc2):
    URL="http://router.project-osrm.org/route/v1/driving/"+loc1+";"+loc2+"?overview=full&alternatives=false&steps=false&geometries=geojson"
    route = "routes"
    distance = "distance"
    duration = "duration"
    geo = "geometry"
    cordinates = "coordinates"
    r = req.get(url=URL)
    data = r.json()
    if(route in data.keys()):
        return data[route][0][duration]
    else:
        -1

def route(loc1,loc2):
    URL="http://router.project-osrm.org/route/v1/driving/"+loc1+";"+loc2+"?overview=full&alternatives=false&steps=false&geometries=geojson"
    route = "routes"
    distance = "distance"
    duration = "duration"
    geo = "geometry"
    cordinates = "coordinates"
    r = req.get(url=URL)
    data = r.json()
    if(route in data.keys()):
        return data[route][0][geo][cordinates]
    else:
        -1
```

```python
data["osrm_eta"] = data.apply(lambda x: route_dis(str(x.pickup_longitude)+","+str(x.pickup_latitude),str(x.dropoff_longitude)+","+str(x.dropoff_latitude)),axis=1)
```

```python
data["route"] = data.apply(lambda x: route(str(x.pickup_longitude)+","+str(x.pickup_latitude),str(x.dropoff_longitude)+","+str(x.dropoff_latitude)),axis=1)
```

```python
data["duration_diff"] = data["trip_duration"]-data["osrm_eta"]
```
